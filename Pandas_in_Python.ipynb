Pandas Series
* A Pandas Series is like a column in a table. It can hold data of any type such as string, integer, float, dictionaries, lists, booleans etc.

```python
import pandas as pd
import numpy as np
```

```python
a = pd.Series([1, 2, 3, 4, 5, 6, 7, 8, np.nan, 97, 90, 100])

print(a)
```

```python
# create labels
a = ([1, 2, 3, 4, 5, 6, 7, 8, 45, 97, 90, 100])

a = pd.Series(a, index = ["apple", "mango", " rice", " yam", " egg", "cat", "dog", " car", "goat", "comb", "water", "soup"])

print(a)
```

```python
# creating a DataFrame
dates = pd.date_range("20240101", periods=12)

dates
```

```python
df = pd.DataFrame(np.random.randn(12, 4), index=dates, columns=list("ABCD"))

df
```

```python
# Create a dataframe
data = {
  "Age": [13, 18, 20, 21, 25, 24],
  "Name": ['Ayo',  'John', 'Gift','Naomi', 'James', 'Joy']
}

#load data into a DataFrame object:
df = pd.DataFrame(data)

print(df)
```

```python
print(df.loc[3]) # locate row
```

```python
print(df.loc[[0, 5]])
```

```python
print(df)
```

```python
df = pd.DataFrame(data, index = ["female", "male", "female", "female", "male", "female"])

print(df)
```
